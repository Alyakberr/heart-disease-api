```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("student-mat.csv", sep=";", quotechar='"')
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

print("Training set shape:", train_set.shape)
print("Test set shape:", test_set.shape)
```

```python
def cap_outliers(series, upper_percentile=0.95):
    cap = series.quantile(upper_percentile)
    return series.apply(lambda x: min(x, cap))

# Apply capping (only to training set to avoid leakage)
train_set["absences"] = cap_outliers(train_set["absences"], 0.95)
train_set["Dalc"] = cap_outliers(train_set["Dalc"], 0.99)
train_set["Walc"] = cap_outliers(train_set["Walc"], 0.99)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Check distributions of all numerical features
numeric_features = train_set.select_dtypes(include=['int64', 'float64']).columns

train_set[numeric_features].hist(bins=20, figsize=(16, 12), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=18)
plt.tight_layout()
plt.show()

# 2. Check skewness
print("Skewness of numerical features:")
skew_vals = train_set[numeric_features].skew().sort_values(ascending=False)
print(skew_vals)

# Optional: visualize skewed features
high_skew = skew_vals[abs(skew_vals) > 1].index.tolist()
print("\nHighly skewed features (|skew| > 1):", high_skew)

# Plot boxplots of skewed features
plt.figure(figsize=(15, 5))
for i, col in enumerate(high_skew):
    plt.subplot(1, len(high_skew), i+1)
    sns.boxplot(y=train_set[col], color="lightcoral")
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()

# 3. Examine value ranges & scales
print("\nValue ranges and scales:")
print(train_set[numeric_features].describe().T[["min", "max", "mean", "std"]])

# 4. Understand categorical-like numeric features (manual review)
ordinal_like_features = ["traveltime", "studytime", "failures", "famrel", "freetime", "goout", "Dalc", "Walc", "health"]
print("\nValue counts for ordinal-like numeric features:")
for col in ordinal_like_features:
    print(f"\n{col} value counts:")
    print(train_set[col].value_counts().sort_index())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# List of skewed features to transform
skewed_features = ["failures", "Dalc", "traveltime", "absences"]

# Apply log1p transformation to these features in the training set
for feature in skewed_features:
    train_set[f"{feature}_log"] = np.log1p(train_set[feature])

# Plot original vs transformed distributions side by side
plt.figure(figsize=(16, 8))

for i, feature in enumerate(skewed_features):
    plt.subplot(2, len(skewed_features), i+1)
    sns.histplot(train_set[feature], kde=True, color='skyblue')
    plt.title(f"Original {feature}")
    
    plt.subplot(2, len(skewed_features), i+1+len(skewed_features))
    sns.histplot(train_set[f"{feature}_log"], kde=True, color='lightgreen')
    plt.title(f"Log-transformed {feature}")

plt.tight_layout()
plt.show()

# Check skewness before and after transformation
print("Skewness before and after log1p transformation:")
for feature in skewed_features:
    original_skew = train_set[feature].skew()
    transformed_skew = train_set[f"{feature}_log"].skew()
    print(f"{feature}: before = {original_skew:.2f}, after = {transformed_skew:.2f}")
```
