### Heart Disease Frequency according to sex

```python
df.sex.value_counts()
```

| Column Name | Description                                          | Value Type               | Example Values                                                                        |
| ----------- | ---------------------------------------------------- | ------------------------ | ------------------------------------------------------------------------------------- |
| `age`       | Age of the patient (in years)                        | Integer                  | 29, 54, 63                                                                            |
| `sex`       | Sex of the patient                                   | Binary                   | 0 = female, 1 = male                                                                  |
| `cp`        | Chest pain type                                      | Categorical (0–3)        | 0 = typical angina<br>1 = atypical angina<br>2 = non-anginal pain<br>3 = asymptomatic |
| `trestbps`  | Resting blood pressure (in mm Hg)                    | Integer                  | 120, 140                                                                              |
| `chol`      | Serum cholesterol (in mg/dl)                         | Integer                  | 240, 230                                                                              |
| `fbs`       | Fasting blood sugar > 120 mg/dl                      | Binary                   | 0 = False, 1 = True                                                                   |
| `restecg`   | Resting electrocardiographic results                 | Categorical (0–2)        | 0 = normal<br>1 = ST-T wave abnormality<br>2 = left ventricular hypertrophy           |
| `thalach`   | Maximum heart rate achieved                          | Integer                  | 150, 180                                                                              |
| `exang`     | Exercise-induced angina                              | Binary                   | 0 = No, 1 = Yes                                                                       |
| `oldpeak`   | ST depression induced by exercise relative to rest   | Float                    | 1.4, 0.6                                                                              |
| `slope`     | Slope of the peak exercise ST segment                | Categorical (0–2)        | 0 = upsloping<br>1 = flat<br>2 = downsloping                                          |
| `ca`        | Number of major vessels (0–3) colored by fluoroscopy | Integer (0–3)            | 0, 1, 2                                                                               |
| `thal`      | Thalassemia                                          | Categorical (1, 2, 3)    | 1 = normal<br>2 = fixed defect<br>3 = reversible defect                               |
| `target`    | Diagnosis of heart disease                           | Binary (target variable) | 0 = No disease<br>1 = Disease                                                         |


```python
#compare target column with sex column
pd.crosstab(df.target, df.sex)
```

```python
#create a plot of cross tab
pd.crosstab(df.target, df.sex).plot(kind="bar",
                                   figsize=(10, 6),
                                   color=["salmon", "lightblue"])
```

### Heart Disease frequency per chest pain type


```python
pd.crosstab(df.cp, df.target)
```

```python
# Cross-tab plot: Chest pain type vs. Target (Heart Disease)
pd.crosstab(df.cp, df.target).plot(kind="bar", 
                                   figsize=(10,6), 
                                   color=["salmon", "lightblue"])

plt.title("Heart Disease Frequency per Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Amount")
plt.legend(["No Disease", "Disease"])  # Corrected this line
plt.xticks(rotation=0)
plt.show()
```
